Import the single project module used throughout the analysis. Import Pandas.

```python
import pandas as pd
import osha_project as osha
```

Load raw Accident and raw Accident Abstract CSV files into Pandas DataFrames. 
Load static Partition CSV file into Pandas DataFrame.

```python
df_acc_original = pd.read_csv('osha_accident.csv', index_col='summary_nr')
df_abs = pd.read_csv('osha_accident_abstract.csv', index_col=['summary_nr','line_nr'])
df_partition = pd.read_csv('partitions.csv', index_col=['summary_nr'])
```

Sort Abstract lines by accident and abstract line number.

```python
df_abs.sort_index(inplace=True)
```

Examine abstract text lines for first 5 accidents.

```python
df_abs[:20]
```

Join multiple abstract text lines for each accident into single text entry. 

```python
joined = df_abs.groupby(level='summary_nr').apply(osha.join_abstract_lines)
df_abs = pd.DataFrame(joined, columns=['summary_txt'])
```

Examine first 5 accidents now with joined abstract text.

```python
df_abs.head()
```

Create new data frame of all three unstructured fields for each accident and filter by accidents in scope for this analysis.

```python
df_txt = df_acc_original[['event_keyword','event_desc']].join(df_partition, how='inner').join(df_abs, how='inner')
df_txt = df_txt[['partition','fatality_ind','event_keyword','event_desc','summary_txt']]
```

Fill the 72 null <i>event_keyword</i> values with <i>NONE</i>

```python
df_txt.event_keyword = df_txt.event_keyword.fillna('NONE')
```

Examine new dataframe metadata.

```python
df_txt
```

Peek at first 5 rows.

```python
df_txt.head()
```

Derive new unstructured fields with selected words strongly correlated with fatality removed.  Remove all 'employee' references as well.

```python
df_txt['event_keyword_clean'] = df_txt['event_keyword'].map(lambda x: osha.clean_keyword(x))
df_txt['event_desc_clean'] = df_txt['event_desc'].map(lambda x: osha.clean_text(x))
df_txt['summary_txt_clean'] = df_txt['summary_txt'].map(lambda x: osha.clean_text(x))
```

Peek at the first 5 rows with the three cleaned text fields added.   

```python
df_txt.head()
```

Save copy of the preprocessed Text data frame.

```python
df_txt.to_csv(osha.project_dir + 'df_txt.csv',index=True)
```

Display list of stemmed tokens highly correlated with fatality to be excluded (and 'employee' stem also).

```python
osha.get_stemmed_exlude_tokens()
```

Retrieve data frames of all tokens and stemmed equivalents for each accident extracted from both the Event Description and Text Summary fields separately.  

```python
df_desc_tokens = osha.get_description_token_table(df_txt)
df_summ_tokens = osha.get_summary_token_table(df_txt)
```

Display stemmed tokens marked for exclusion, all word variants with the stemmed token as root, and occurrence counts detected within the Event Description field across all accidents.

```python
df_desc_tokens[df_desc_tokens.excluded==1].groupby(['stem','token']).size()
```

Display stemmed tokens marked for exclusion, all word variants with the stemmed token as root, and occurrence counts detected within the Text Summary field across all accidents.

```python
df_summ_tokens.groupby(['stem','token']).size()
```

Generate set of all word variants from the Event Description and Text Summary fields to be removed from text fields during preprocessing.

```python
exclude_words = set(list(df_desc_tokens[df_desc_tokens.excluded==1].token.unique())+list(df_summ_tokens.token.unique()))
exclude_words
```

Create pivot tables for subsequent analysis.

```python
desc_piv = df_desc_tokens.pivot_table('token',rows=['stem','excluded'],cols='fatality_ind',aggfunc='count').fillna(0)
desc_piv.columns=['non_fatal','fatal']
desc_piv = desc_piv.reset_index().set_index('stem')
desc_piv['tot_occur']=desc_piv.fatal + desc_piv.non_fatal
desc_piv['pct_fatal']=desc_piv.fatal/desc_piv.tot_occur

summ_piv = df_summ_tokens.pivot_table('token',rows=['stem','excluded'],cols='fatality_ind',aggfunc='count').fillna(0)
summ_piv.columns=['non_fatal','fatal']
summ_piv = summ_piv.reset_index().set_index('stem')
summ_piv['tot_occur']=summ_piv.fatal + summ_piv.non_fatal
summ_piv['pct_fatal']=summ_piv.fatal/summ_piv.tot_occur
```

Examine statistics of excluded stemmed tokens from the Event Description field.

```python
desc_piv[desc_piv.excluded==1].sort('tot_occur',ascending=False)
```

Examine statistics of excluded stemmed tokens from the Text Summary field.

```python
summ_piv[summ_piv.excluded==1].sort('tot_occur',ascending=False)
```
