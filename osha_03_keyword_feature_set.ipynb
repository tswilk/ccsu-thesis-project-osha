Import the single project module used throughout the analysis. Import Pandas.

```python
import pandas as pd
import osha_project as osha
```

Load the preprocessed text CSV file into Pandas DataFrame and create a new data frame with cleaned keywords only.

```python
df_txt = pd.read_csv(osha.project_dir + 'df_txt.csv', index_col='summary_nr')
df_kw = df_txt[['partition','fatality_ind','event_keyword_clean']]
```

Peek at first 5 records.

```python
df_kw.head()
```

Iterate through all accidents of the train partition, identify set of unique keywords, and create a new data frame of zeroes with each unique keyword as a column. 

```python
keyword_iter_train = (set(str(x).split(',')) for x in osha.get_partition(df_kw,'train').event_keyword_clean)
keywords_train = sorted(set.union(*keyword_iter_train))

kw_dummies = pd.DataFrame(np.zeros((len(df_kw),len(keywords_train))), columns = keywords_train, \
                          index = df_kw.index).reset_index()
```

Iterate through all accident keywords and assign a one to each keyword column associated with each accident.

```python
for i, gen in enumerate(df_kw.event_keyword_clean):
    kws = str(gen).split(',')
    for kw in kws:
        if kw in keywords_train:
            kw_dummies.ix[i,kw] = 1
```

Create Keyword feature set for modeling.

```python
kw_dummies = kw_dummies.set_index('summary_nr').rename(columns=lambda x: 'kw_' + x.replace(' ','_'))
df_kw = df_kw[['partition','fatality_ind']].join(kw_dummies)
```

Select top 25 features based on the train set and plot distribution of top features by partition. 

```python
df_top_features = osha.get_top_k_features(df_kw, osha.f_classif, 25)
df = df_kw[['fatality_ind','partition']+list(df_top_features.feature)]
ax = df[df.partition != 'validation'].groupby(['partition']).mean().T[:25].plot(kind='barh')
ax.set_xlabel('Average Indicator Value')
ax.set_title('Distribution of Top 25 Keyword Features by Partition')
```

Initialize K-means model with 4 clusters and fit model with the train partition data.

```python
k_means = osha.KMeans(init='k-means++', n_clusters=4, n_init=10)
k_means.fit(osha.get_partition(df_kw,'train').drop(['fatality_ind','partition'],axis=1))
```

Add new field to Keyword feature set designating cluster membership.  

```python
preds = k_means.predict(df_kw.drop(['partition','fatality_ind'],axis=1))
df_kw['kw_kmeans_cluster'] = preds
```

Create data frame for analysis comprised of the fatality indicator, partition, top 25 keyword features and cluster membership.

```python
df_kw_cluster = df_kw[df_kw.partition != 'validation'][['partition','fatality_ind','kw_kmeans_cluster']+\
                                                        list(df_top_features.feature)]
```

Plot graph of K-Means cluster counts by partition.

```python
df_kw_cluster.groupby(['partition','kw_kmeans_cluster']).size().unstack().\
                       plot(kind='bar',legend=False, title='K-Means Cluster Counts')
```

Plot mean keyword frequency of top 25 keyword features from the train partition.

```python
ax = osha.get_partition(df_kw_cluster, 'train').groupby(['kw_kmeans_cluster']).mean().T.\
                   plot(kind='bar', sharey=True, figsize=(10,10), \
                   title='Mean Keyword Frequency by K-Means Cluster - Train Set', subplots=True)
```

Plot mean keyword frequency of top 25 keyword features from the test partition.

```python
osha.get_partition(df_kw_cluster, 'test').groupby(['kw_kmeans_cluster']).mean().T.\
                   plot(kind='bar', sharey=True, figsize=(10,10), \
                   title='Mean Keyword Frequency by K-Means Cluster - Test Set', subplots=True)
```

Derive dummy indicators for cluster membership in preparation for modeling, and view first 5 records.

```python
dummies = pd.get_dummies(preds)
dummies.columns=['kmeans_cluster_' + str(x) for x in range(4)]
dummies.index = df_kw.index
dummies.head()
```

Join cluster membership dummy indicators to Keyword feature set.

```python
df_kw = df_kw.drop(['kw_kmeans_cluster'],axis=1)
df_kw = df_kw.join(dummies)
```

Save a copy of the Keyword feature set now ready for modeling.

```python
df_kw.to_csv(osha.keyword_dir + 'df_kw.csv',index=True)
```

Set Keyword feature set variables.

```python
fs_dir = osha.keyword_dir
fs_name = 'Keyword'
```

Train classification models and log results and metrics for subsequent analysis.

Load logged results and metrics data of the best Structured classification model for comparison (Logistic Regression base model).

```python
df_results_struct = osha.pickle.load(open(osha.structured_dir + 'df_results.p','rb'))
df_results_struct = df_results_struct[df_results_struct.pair_name=='Structured - LogReg']
df_metrics_struct = osha.pickle.load(open(osha.structured_dir + 'df_metrics.p','rb'))
df_metrics_struct = df_metrics_struct[df_metrics_struct.pair_name=='Structured - LogReg']
```

Load trained base classification model results, compute combination model results and extend results in preparation for subsequent analyses.

```python
df_results, df_metrics, col_dict, model_dict, fsets, clfs = osha.get_model_training_info(fs_dir)
df_mrp_results = osha.get_mrp_results(df_results)
df_voting_results = osha.get_voting_results(df_results, len(clfs))
df_mrp_metrics = osha.get_combination_metrics(df_mrp_results)
df_voting_metrics = osha.get_combination_metrics(df_voting_results)
df_results = df_results.append(df_results_struct)
df_metrics = df_metrics.append(df_metrics_struct)
df_all_results = df_results.append(df_mrp_results).append(df_voting_results)
df_all_metrics = df_metrics.append(df_mrp_metrics).append(df_voting_metrics)
df_all_results['target_hit'] = 1 - abs(df_all_results.target_ind - df_all_results.target_pred)
outcome_nbr_map = {'TP':-2 ,'FN':-1, 'FP':1, 'TN':2}
df_all_results['outcome'] = df_all_results[['target_ind','target_pred']].apply(osha.get_outcome,axis=1)
df_all_results['outcome_nbr'] = df_all_results['outcome'].map(outcome_nbr_map)
df_all_results['model_type'] = df_all_results['model_name'].apply(lambda x: osha.get_model_type(x))
df_all_metrics['model_type'] = df_all_metrics['model_name'].apply(lambda x: osha.get_model_type(x))
```

Load trained base classification model results, compute combination model results and extend results in preparation for subsequent analyses.

```python
df_base_metrics = df_all_metrics[df_all_metrics.model_type=='base']
df_base_results = df_all_results[df_all_results.model_type=='base']
struct_FN_idx = df_results_struct[(df_results_struct.target_ind==1)&(df_results_struct.target_pred==0)].index
struct_FP_idx = df_results_struct[(df_results_struct.target_ind==0)&(df_results_struct.target_pred==1)].index
df_piv_pred_0_prob = df_base_results.reset_index().pivot('summary_nr','pair_name','pred_0_prob')
df_piv_pred_1_prob = df_base_results.reset_index().pivot('summary_nr','pair_name','pred_1_prob')
df_struct_FN = df_piv_pred_1_prob.ix[struct_FN_idx].sort('Structured - LogReg',ascending=True)
df_struct_FP = df_piv_pred_0_prob.ix[struct_FP_idx].sort('Structured - LogReg',ascending=True)
df_outcome = df_base_results.reset_index().pivot('summary_nr','pair_name','outcome_nbr').\
    join(df_results_struct.pred_1_prob,how='inner')
df_outcome.sort('pred_1_prob',ascending=False,inplace=True)
df_outcome = df_outcome.drop('pred_1_prob',axis=1)
```
