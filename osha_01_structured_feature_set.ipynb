Load raw Accident and Injury CSV files into Pandas DataFrames. 
Load Code to Value mapping CSV file into Pandas DataFrame.

```python
df_acc = pd.read_csv('osha_accident.csv', index_col='summary_nr')
df_inj = pd.read_csv('osha_accident_injury.csv', index_col=['summary_nr'])
df_map = pd.read_csv('osha_code_map.csv')
```

```python
df_acc
```

Transform <i>fatality</i> field into a 1/0 indicator. This is the target variable used throughout the analysis. Drop original. 

```python
df_acc['fatality_ind'] = df_acc['fatality'].replace(['X',np.nan],[1,0]).astype(int64)
df_acc = df_acc.drop(['fatality'],axis=1)
```

Drop semi-structured and unstructured fields. Will use in later steps.

```python
df_acc = df_acc.drop(['event_keyword','event_desc','abstract_text'],axis=1)
```

Drop fields with no values present (2) and identifier fields not necessary for this analysis (1).

```python
df_acc = df_acc.drop(['event_time','state_flag','report_id'],axis=1)
```

Drop Construction specific fields not used for this analysis.

```python
df_acc = df_acc.drop(['const_end_use','build_stories','nonbuild_ht','project_cost','project_type'],axis=1)
```

Fill SIC Code cells with missing values (13) and multiple designations with -1 to signal missing or multiple.  Drop original.

```python
df_acc.sic_list = df_acc.sic_list.fillna(-1)
df_acc['sic_code'] = df_acc['sic_list'].map(lambda x: int(x) if len(str(x).strip()) == 4 else -1)
df_acc = df_acc.drop(['sic_list'],axis=1)
```

Derive time-based fields from <i>event_date</i>. Drop original.

```python
df_acc['event_ts'] = df_acc['event_date'].map(osha.parser.parse)
df_acc['event_year'] = df_acc['event_ts'].map(lambda x: x.year)
df_acc['event_month'] = df_acc['event_ts'].map(lambda x: x.month)
df_acc['event_weekday'] = df_acc['event_ts'].map(lambda x: x.dayofweek)
df_acc['event_hour'] = df_acc['event_ts'].map(lambda x: x.hour)
df_acc = df_acc.drop(['event_date'],axis=1)
```

Create table of fatality count and injury count by accident, bin injury count, and view first 5 rows.

```python
num_injuries = pd.Series(df_inj.groupby(level='summary_nr').size(),name='num_injuries')
num_fatalities = pd.Series(df_inj.groupby(level='summary_nr').fatality_ind.sum(),name='num_fatalities')
num_inj_fatalities = pd.DataFrame({'num_injuries':num_injuries, \
                                   'num_fatalities':num_fatalities},index=num_injuries.index).reset_index()
cats = pd.cut(num_inj_fatalities.num_injuries, range(1,13)+[153], right=False)
num_inj_fatalities['bin_num_injuries'] = cats
num_inj_fatalities['bin_num_injuries_order'] = cats.labels
num_inj_fatalities.head()
```

Create table of mean fatalities and accident count by binned injury count, and view table. 

```python
num_inj_fatalities_mean = num_inj_fatalities.groupby(['bin_num_injuries_order','bin_num_injuries']).num_fatalities.mean()
num_inj_size = num_inj_fatalities.groupby(['bin_num_injuries_order','bin_num_injuries']).size()
inj_fin = pd.DataFrame({'num_inj_fatalities_mean':num_inj_fatalities_mean, \
                        'num_inj_size':num_inj_size},index=num_inj_size.index).reset_index()
inj_fin
```

Map codes to values for selected categorical fields.

```python
struct_cols_to_map = ['sic_code', 'nature_of_inj', 'part_of_body', 'src_of_injury', 'event_type',
                      'evn_factor', 'hum_factor', 'occ_code', 'task_assigned']

for col in struct_cols_to_map:
    df_col = df_map[df_map.field_name==col]
    mapping_dict = {code:value for code,value in zip(list(df_col.code_value),list(df_col.mapped_value))}
    df_struct[col]=df_struct[col].map(mapping_dict)
```

Convert categorical fields into dummy indicators in preparation for modeling.

```python
dummies_list = []
struct_one_hot_vars = ['sic_desc', 'event_hour', 'nature_of_inj', 'part_of_body', 'src_of_injury','event_type',\
                       'evn_factor', 'hum_factor']

for one_hot_var in struct_one_hot_vars:
    dummies = pd.get_dummies(df_struct[one_hot_var],prefix=one_hot_var)
    dummies_list.append(dummies)
df_dummies = pd.concat(dummies_list,axis=1)
```

Drop converted categorical fields and join dummy indicator equivalents.

```python
df_struct = df_struct.drop(struct_one_hot_vars,axis=1).join(df_dummies)
```

Load trained base classification model results, compute combination model results and extend results in preparation for subsequent analyses.

```python
df_results, df_metrics, col_dict, model_dict, fsets, clfs = osha.get_model_training_info(fs_dir)
df_mrp_results = osha.get_mrp_results(df_results)
df_voting_results = osha.get_voting_results(df_results, len(clfs))
df_mrp_metrics = osha.get_combination_metrics(df_mrp_results)
df_voting_metrics = osha.get_combination_metrics(df_voting_results)
df_all_results = df_results.append(df_mrp_results).append(df_voting_results)
df_all_metrics = df_metrics.append(df_mrp_metrics).append(df_voting_metrics)
df_all_results['target_hit'] = 1 - abs(df_all_results.target_ind - df_all_results.target_pred)
outcome_nbr_map = {'TP':-2 ,'FN':-1, 'FP':1, 'TN':2}
df_all_results['outcome'] = df_all_results[['target_ind','target_pred']].apply(osha.get_outcome,axis=1)
df_all_results['outcome_nbr'] = df_all_results['outcome'].map(outcome_nbr_map)
df_all_results['model_type'] = df_all_results['model_name'].apply(lambda x: osha.get_model_type(x))
df_all_metrics['model_type'] = df_all_metrics['model_name'].apply(lambda x: osha.get_model_type(x))
```
