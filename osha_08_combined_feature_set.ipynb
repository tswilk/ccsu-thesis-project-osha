Import the single project module used throughout the analysis. Import Pandas.

```python
import pandas as pd
import osha_project as osha
```

Load the Main feature sets.

```python
main_fsets={}
main_fsets['Keyword'] = pd.read_csv(osha.keyword_dir + 'df_kw.csv', index_col='summary_nr')
main_fsets['Topic'] = pd.read_csv(osha.topic_dir + 'df_topic.csv', index_col='summary_nr')
main_fsets['Linguistic'] = pd.read_csv(osha.linguistic_dir + 'df_ling.csv', index_col='summary_nr')
main_fsets['Structured'] = pd.read_csv(osha.structured_dir + 'df_struct.csv', index_col='summary_nr')
```

Load the selected SVD feature sets.

```python
svd_fsets={}
svd_fsets['Desc_SVD'] = pd.read_csv(osha.svd_description_dir + 'df_desc_stem_n1.csv', index_col='summary_nr')
svd_fsets['Summ_SVD'] = pd.read_csv(osha.svd_summary_dir + 'df_summ_stem_n1.csv', index_col='summary_nr')
```

Consolidate the Main feature sets and selected SVD feature sets into All feature sets.

```python
all_fsets={}
all_fsets.update(main_fsets)
all_fsets.update(svd_fsets)
```

Create 28 variations of Combined feature sets for modeling.

```python
combined_fsets={}
combined_fsets['Main Sep P05'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='pct',ptile=5)
combined_fsets['Main Sep P10'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='pct',ptile=10)
combined_fsets['Main Sep P20'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='pct',ptile=20)
combined_fsets['Main Sep K10'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='kbest',nFeatures=10)
combined_fsets['Main Sep K05'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='kbest',nFeatures=5)
combined_fsets['Main Sep K02'] = osha.get_combined_separate_fsets(main_fsets,fs_fn='kbest',nFeatures=2)
combined_fsets['Main Utd P01'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=1)
combined_fsets['Main Utd P02'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=2)
combined_fsets['Main Utd P05'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=5)
combined_fsets['Main Utd P10'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=10)
combined_fsets['Main Utd P20'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=20)
combined_fsets['Main Utd P30'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=30)
combined_fsets['Main Utd P40'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=40)
combined_fsets['Main Utd P50'] = osha.get_combined_united_fsets(main_fsets,fs_fn='pct',ptile=50)
combined_fsets['All Sep P05'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='pct',ptile=5)
combined_fsets['All Sep P10'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='pct',ptile=10)
combined_fsets['All Sep P20'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='pct',ptile=20)
combined_fsets['All Sep K10'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='kbest',nFeatures=10)
combined_fsets['All Sep K05'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='kbest',nFeatures=5)
combined_fsets['All Sep K02'] = osha.get_combined_separate_fsets(all_fsets,fs_fn='kbest',nFeatures=2)
combined_fsets['All Utd P01'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=1)
combined_fsets['All Utd P02'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=2)
combined_fsets['All Utd P05'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=5)
combined_fsets['All Utd P10'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=10)
combined_fsets['All Utd P20'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=20)
combined_fsets['All Utd P30'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=30)
combined_fsets['All Utd P40'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=40)
combined_fsets['All Utd P50'] = osha.get_combined_united_fsets(all_fsets,fs_fn='pct',ptile=50)
```

Examine classifiers and default parameter settings.

```python
clfs = osha.get_classifiers()
clfs
```

Set Combined feature sets variable.

```python
fs_dir = osha.combined_dir
```

Train classification models and log results and metrics for subsequent analysis.

```python
osha.process_model_batch(combined_fsets, clfs, fs_dir)
```

Load logged results and metrics data of the best Structured classification model for comparison (Logistic Regression base model).

```python
df_results_struct = osha.pickle.load(open(osha.structured_dir + 'df_results.p','rb'))
df_results_struct = df_results_struct[df_results_struct.pair_name=='Structured - LogReg']
df_metrics_struct = osha.pickle.load(open(osha.structured_dir + 'df_metrics.p','rb'))
df_metrics_struct = df_metrics_struct[df_metrics_struct.pair_name=='Structured - LogReg']
```

Load trained base classification model results, compute combination model results and extend results in preparation for subsequent analyses.

```python
df_results, df_metrics, col_dict, model_dict, fsets, clfs = osha.get_model_training_info(fs_dir)
df_mrp_results = osha.get_mrp_results(df_results)
df_voting_results = osha.get_voting_results(df_results, len(clfs))
df_mrp_metrics = osha.get_combination_metrics(df_mrp_results)
df_voting_metrics = osha.get_combination_metrics(df_voting_results)
df_results = df_results.append(df_results_struct)
df_metrics = df_metrics.append(df_metrics_struct)
df_all_results = df_results.append(df_mrp_results).append(df_voting_results)
df_all_metrics = df_metrics.append(df_mrp_metrics).append(df_voting_metrics)
df_all_results['target_hit'] = 1 - abs(df_all_results.target_ind - df_all_results.target_pred)
outcome_nbr_map = {'TP':-2 ,'FN':-1, 'FP':1, 'TN':2}
df_all_results['outcome'] = df_all_results[['target_ind','target_pred']].apply(osha.get_outcome,axis=1)
df_all_results['outcome_nbr'] = df_all_results['outcome'].map(outcome_nbr_map)
df_all_results['model_type'] = df_all_results['model_name'].apply(lambda x: osha.get_model_type(x))
df_all_metrics['model_type'] = df_all_metrics['model_name'].apply(lambda x: osha.get_model_type(x))
```

Display base model only results ranked by accuracy score, high to low.

```python
df_metrics.sort('accuracy_score',ascending=False)[:30]
```

Display base model, mean response probability model and voting model results ranked by accuracy score, high to low.

```python
df_all_metrics.sort('accuracy_score',ascending=False)[:30]
```

Select subset of combined feature sets for further analysis and filter results data.

```python
selected_fsets = ['All Utd P40','All Utd P05','All Sep P20','Main Utd P30','Main Sep P20','Structured']
df_results = df_results[df_results.fset_name.isin(selected_fsets)]
df_metrics = df_metrics[df_metrics.fset_name.isin(selected_fsets)]
df_all_results = df_all_results[df_all_results.fset_name.isin(selected_fsets)]
df_all_metrics = df_all_metrics[df_all_metrics.fset_name.isin(selected_fsets)]
```

Plot feature importance for the top 20 features of the Random Forest classification models.

```python
df_tree = osha.get_tree_model_info(selected_fsets, clfs, model_dict, col_dict)
osha.plot_tree_importance(selected_fsets, clfs, 20, model_dict, col_dict, fs_dir, save=True)
```

Plot feature importance for the top 20 features of the Logistic Regression classification models.

```python
df_logreg = osha.get_logreg_model_info(selected_fsets, clfs, model_dict, col_dict)
osha.plot_logreg_coef_importance(selected_fsets, clfs, df_logreg, 10, model_dict, col_dict, fs_dir, save=True)
```

Plot gains chart for the base classification models.

```python
osha.plot_gains_charts(df_results, clfs, selected_fsets, 'fset', fs_dir,\
                       save=True, dfs=df_results_struct, dfsn='Best Structured Model')
```

Plot mean response probability model and voting model results chart.

```python
osha.plot_combination_model_chart(df_mrp_metrics, df_voting_metrics, selected_fsets, len(clfs), fs_dir, save=False)
```

Load trained base classification model results, compute combination model results and extend results in preparation for subsequent analyses.

```python
df_base_metrics = df_all_metrics[df_all_metrics.model_type=='base']
df_base_results = df_all_results[df_all_results.model_type=='base']
struct_FN_idx = df_results_struct[(df_results_struct.target_ind==1)&(df_results_struct.target_pred==0)].index
struct_FP_idx = df_results_struct[(df_results_struct.target_ind==0)&(df_results_struct.target_pred==1)].index
df_piv_pred_0_prob = df_base_results.reset_index().pivot('summary_nr','pair_name','pred_0_prob')
df_piv_pred_1_prob = df_base_results.reset_index().pivot('summary_nr','pair_name','pred_1_prob')
df_struct_FN = df_piv_pred_1_prob.ix[struct_FN_idx].sort('Structured - LogReg',ascending=True)
df_struct_FP = df_piv_pred_0_prob.ix[struct_FP_idx].sort('Structured - LogReg',ascending=True)
df_outcome = df_base_results.reset_index().pivot('summary_nr','pair_name','outcome_nbr').\
    join(df_results_struct.pred_1_prob,how='inner')
df_outcome.sort('pred_1_prob',ascending=False,inplace=True)
df_outcome = df_outcome.drop('pred_1_prob',axis=1)
```

Plot outcome comparison (TP,FP,FN,TN) for best Structured classification model and all 10 Combined classification models. 

```python
osha.plot_outcome_comparison(df_outcome, fs_dir, figsize=(10,6), save=True)
```

Plot best Structured classification model False Negative cases along with all 10 Combined classification model outcomes.

```python
osha.plot_best_FN(df_struct_FN, fs_dir, save=True)
```

Plot best Structured classification model False Positive cases along with all 10 Combined classification model outcomes.

```python
osha.plot_best_FP(df_struct_FP, fs_dir, save=True)
```
